```python
import pandas as pd

df = pd.read_csv('soil_fertilization.csv')

print(df.corr()['Output'])
df = df[(df['Output'] == 1) | (df['Output'] == 0)]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report


scaler = StandardScaler()
dt = DecisionTreeClassifier(max_depth=5,random_state=42, class_weight='balanced')
rf = RandomForestClassifier(max_depth=5, random_state=42, class_weight='balanced')

x = df.drop(columns='Output')
y = df['Output']

x_train,x_test,y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=42)

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

dt.fit(x_train, y_train)
rf.fit(x_train, y_train)

y_dt = dt.predict(x_test)
y_rf = rf.predict(x_test)

dtr = classification_report(y_test, y_dt)
rfr= classification_report(y_test, y_rf)

print(dtr)
print(rfr)

```
